```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

#  Random Walk

##  1D Uniforme

```python
step = -1,1
N = 100000
########## Passo
X = 0
NormX = 0
UniformX = 0
########## Caminho
BinaryWalk = []
NormWalk = []
UniformWalk = []
##########

for i in range(0,N):
    BinaryWalk.append(X)
    X = X +random.choice(step)
    NormWalk.append(NormX)
    NormX = NormX +random.normalvariate(mu=0,sigma=0.33)
    UniformWalk.append(UniformX)
    UniformX = UniformX + random.uniform(-1,1)
BinaryWalk = np.array(BinaryWalk)
NormWalk = np.array(NormWalk)
UniformWalk = np.array(UniformWalk)

print(f'Binary Mean: {BinaryWalk.mean()} -- Binary Standard Deviantion: {BinaryWalk.std()}')
print(f'Normal Distribution Mean: {NormWalk.mean()} --Normal Standard Deviation: {NormWalk.std()}')
print(f'Uniform Mean: {UniformWalk.mean()} -- Uniform Standard Deviation: {UniformWalk.std()}')

fig = plt.figure(figsize=(10,5),dpi = 100)
plt.title(f'RandomWalk 1D com N = {N}')
plt.plot(BinaryWalk,label = 'Binary Choosing')
plt.plot(NormWalk, label ='Normal Distribution')
plt.plot(UniformWalk,label = 'Uniform Distribution')
plt.legend()
plt.show()
```

## 2D Binário

```python
step = -1,1
N = 1000000
X = 0
Y = 0
RdnWlkX = []
RdnWlkY = []
for i in range(0,N):
    RdnWlkX.append(X)
    RdnWlkY.append(Y)
    X = X +random.choice(step)
    Y = Y +random.choice(step)
RdnWlkX = np.array(RdnWlkX)
RdnWlkY = np.array(RdnWlkY)
fig = plt.figure(figsize=(8,8),dpi = 100)
plt.title(f'RandomWalk 2D com N = {N}')
plt.plot(RdnWlkX,RdnWlkY)
plt.show()
```

## 2D RandomWalk Uniform Distribution (-1,1)

```python
N = 100000
X = 0
Y = 0
RdnWlkX = []
RdnWlkY = []
for i in range(0,N):
    RdnWlkX.append(X)
    RdnWlkY.append(Y)
    X = X +random.uniform(-1,1)
    Y = Y +random.uniform(-1,1)
RdnWlkX = np.array(RdnWlkX)
RdnWlkY = np.array(RdnWlkY)
print(f'Mean X: {RdnWlkX.mean()} ----Standard Deviation X: {RdnWlkX.std()}')
print(f'Mean Y: {RdnWlkY.mean()} ----Standard Deviation Y: {RdnWlkY.std()}')
fig = plt.figure(figsize=(8,8),dpi = 100)
plt.title(f'RandomWalk 2D Uniform Distribuition com N = {N}')
plt.plot(RdnWlkX,RdnWlkY)
plt.show()
```

## 3D Binary

### Walking Generator

```python
step = -1,1
N = 100000
#####
X = 0
Y = 0
Z = 0
RdnWlkX = []
RdnWlkY = []
RdnWlkZ = []
for i in range(0,N):
    RdnWlkX.append(X)
    RdnWlkY.append(Y)
    RdnWlkZ.append(Z)
    X = X +random.choice(step)
    Y = Y +random.choice(step)
    Z = Z +random.choice(step)
    
RdnWlkX = np.array(RdnWlkX)
RdnWlkY = np.array(RdnWlkY)
RdnWlkZ = np.array(RdnWlkZ)
```

### Plotagem 3D com Rotação

```python
ax = plt.figure(figsize=(8,8),dpi = 100).add_subplot(projection = '3d')
plt.title(f'RandomWalk 3D com N = {N}')
ax.view_init(30,30)
plt.plot(RdnWlkX,RdnWlkY,RdnWlkZ,label = 'Binary Walk')

plt.legend()
plt.show()
```

## 3D Normal Distribution

### Walking Generator

```python
N = 100000
NormX = 0
NormY = 0
NormZ = 0
NormalWalkX = []
NormalWalkY = []
NormalWalkZ = []
for i in range(0,N):
    NormalWalkX.append(NormX)
    NormalWalkY.append(NormY)
    NormalWalkZ.append(NormZ)
    NormX = NormX +random.normalvariate(mu=0,sigma=0.33)
    NormY = NormY +random.normalvariate(mu=0,sigma=0.33)
    NormZ = NormZ +random.normalvariate(mu=0,sigma=0.33)
NormalWalkX = np.array(NormalWalkX)
NormalWalkY = np.array(NormalWalkY)
NormalWalkZ = np.array(NormalWalkZ)
```

### Plotting and Rotation

```python
ax = plt.figure(figsize=(8,8),dpi = 100).add_subplot(projection = '3d')
plt.title(f'RandomWalk 3D com N = {N}')
###Select the Observation Angle
ax.view_init(30,30)
plt.plot(NormalWalkX,NormalWalkY,NormalWalkZ,label = 'Normal Walk')

plt.legend()
plt.show()
```
